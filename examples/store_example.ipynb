## Setup Qdrant Instance

    1. Setup Local Qdrant Instance with Docker Container

- Create a docker-compose.yml file at the root of your project and copy the yaml code below into it.
- This section assumes you have installed `Docker Desktop` for `Windows/Mac` or `Docker Engine` for `Linux` machines.

```yaml
services:
  qdrant:
    image: qdrant/qdrant:v1.18.0
    container_name: qdrant
    restart: unless-stopped

    ports:
      - "6333:6333"
      - "6334:6334"

    volumes:
      - ./data/qdrant:/qdrant/storage

    environment:
      QDRANT__SERVICE__API_KEY: ${QDRANT_API_KEY}

volumes: {}
```

    2.   Create a .venv file at the root of your project and copy the content below into it.

QDRANT_API_KEY=change-me
QDRANT_URL=http://localhost:6333

- Replace the value of QDRANT_API_KEY with your own value
- If you are using cloud qdrant instance, replace the QDRANT_URL with the url for your cloud instance

## Download Sample Dataset

````python
from pathlib import Path

from datasets import load_dataset


def get_dataset(
    dataset_path="virattt/financial-qa-10K",
    save_path="./data/financial_qa_10k.parquet",
    use_saved=True,
):
    if use_saved and Path(save_path).exists():
        dataset = load_dataset("parquet", data_files=save_path)
        return dataset["train"]

    # Load dataset from hugging face and save to disk
    dataset = load_dataset(dataset_path, split="train")
    Path("./data").mkdir(parents=True, exist_ok=True)
    dataset.to_parquet(save_path)

    return dataset
````
